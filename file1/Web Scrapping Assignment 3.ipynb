```python
search_bar=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
search_bar

search_bar.send_keys(enter)

button=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input")
button.click()
```

```python
##Q2
```

```python
Brand = []
Price = []
Expected_Delivery = []
Product_url = []
```

```python
start = 0
end = 2
for page in range(start,end):
    brand_tag = driver.find_elements(By.XPATH,'//span[@class="a-size-medium a-color-base a-text-normal"]')
    for i in brand_tag:
        Brand.append(i.text)
        
    price_tag = driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
    for i in price_tag:
        Price.append(i.text)
        
    ed_tag = driver.find_elements(By.XPATH,'//span[@class="a-color-base a-text-bold"]')
    for i in ed_tag:
        Expected_Delivery.append(i.text)
        
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        Product_url.append(i.get_attribute('href'))
    
```

```python
import pandas as pd
df = pd.DataFrame({'Brand':Brand,'Price':Price,'Expected Delivery':Expected_Delivery,'URL':Product_url})
df
```

```python
##Q5  Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps. 
```

```python
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import requests
import re
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.google.co.in/maps")
```

```python
city = input('Enter City Name : ')                                        
search = driver.find_element(By.XPATH,'//input[@class="searchboxinput xiQnY"]')                    
search.clear()
search.send_keys(city) 
button = driver.find_element(By.XPATH,'//button[@class="mL3xi google-symbols"]') 
button.click()

```

```python
try:
    url_string = driver.current_url
    print("URL Extracted: ", url_string)
    lat_lng = re.findall(r'@(.*)data',url_string)
    if len(lat_lng):
        lat_lng_list = lat_lng[0].split(",")
        if len(lat_lng_list)>=2:
            lat = lat_lng_list[0]
            lng = lat_lng_list[1]
        print("Latitude = {}, Longitude = {}".format(lat, lng))

except Exception as e:
        print("Error: ", str(e))
```
